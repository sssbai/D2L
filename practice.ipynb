```python
# Gradient calculation under control flow
import torch
def f(a):
    b = a * 2
    while b.norm() < 1000:
        b = b * 2
    if b.sum() > 0:
        c = b
    else:
        c = 100 * b
    return c
a = torch.randn(1, requires_grad=True)
d = f(a)
print(d)
d.backward()
a.grad, a.grad == d/a
```
